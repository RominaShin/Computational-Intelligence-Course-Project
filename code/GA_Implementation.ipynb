```python
import pandas as pd
import random
from collections import Counter
```

```python
production_df = pd.read_csv('Asia - Production_processed.csv')
export_df = pd.read_csv('Asia - Export_processed.csv')
```

```python
def set_year(df, last_year):
  for year in df['Year']:
    if year > last_year:
      df = df[~df['Year'].isin([year])]
  return df
```

```python
items = [item for item in production_df.drop_duplicates(subset=['Item'])['Item']]
countries = [country for country in production_df.drop_duplicates(subset=['Area'])['Area']]
```

```python
total_values = dict()
production_df = set_year(production_df, 2010)
for country in countries:
  total_values.update({country: production_df['Value'][production_df['Area'] == country].sum()})
print(list({k: v for k, v in sorted(total_values.items(), key=lambda item: item[1], reverse=True)}))
```
